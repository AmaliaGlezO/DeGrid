Picos de Demanda: En lugar de usar ondas sinusoidales simples, el script utiliza Campanas de Gauss. Matemáticamente, la fórmula es $f(x) = A \cdot e^{-0.5 \cdot (\frac{x-\mu}{\sigma})^2}$. Esto permite crear un "pico" muy localizado a una hora específica ($\mu$), como las 20:00 hrs, sin afectar el resto del día.Comportamiento de Plantas: * La Solar tiene una campana invertida al mediodía.La Termoeléctrica es una línea casi recta que absorbe la carga hasta llegar a su límite técnico.Los Bloques Flotantes actúan como una capa de emergencia; sus valores serán 0.0 la mayor parte de la madrugada, pero se activarán automáticamente durante los picos de calor o en el horario estelar cuando la termoeléctrica y la solar no sean suficientes.
